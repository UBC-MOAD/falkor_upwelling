```python
filt_csts1 = TSmode.filter_CTD_list('YEAR',2008)
filt_csts = []
for ii in filt_csts1:
    if ii['Month'] >= 6 and ii['Month'] <= 8:
        filt_csts.append(ii)

fig,ax = plt.subplots()        
m = Basemap(llcrnrlon=-130,llcrnrlat=35,urcrnrlon=-122,urcrnrlat=52,projection='mill')
lon,lat = m([k['Longitude'] for k in filt_csts],[k['Latitude'] for k in filt_csts])
points = ax.plot(lon,lat,'sk',ms=4)

dat = pd.DataFrame(index=range(len(filt_csts)))
dat['lon'] = np.array([k['Longitude'] for k in filt_csts])
dat['lat'] = np.array([k['Latitude'] for k in filt_csts])

labels = []
for ii in range(len(filt_csts)):
    label = dat.ix[[ii], :].T
    label.columns = ['Cast {0}'.format(ii)]
    labels.append(str(label.to_html()))

# NOTE: Following code adapted from mpld3 example 'HTML Tooltips'
# Define some CSS to control our custom labels
css = """
table
{
  border-collapse: collapse;
}
th
{
  color: #ffffff;
  background-color: #000000;
}
td
{
  background-color: #cccccc;
}
table, th, td
{
  font-family:Arial, Helvetica, sans-serif;
  border: 1px solid black;
  text-align: right;
}
"""
tooltip = plugins.PointHTMLTooltip(points[0], labels,
                                   voffset=10, hoffset=10, css=css)
plugins.connect(fig, tooltip)

#csts = [805,854,907,653,716,624,1016,1014,375,770,824,391,412,433,448,798,463,468]
#line_lon,line_lat = m([filt_csts[x]['Longitude'] for x in csts],[filt_csts[x]['Latitude'] for x in csts])
#plt.plot(line_lon,line_lat,'-r')

csts = [561,537,461,524,432,608,202,225,246,254,265,276,287,312]
line_lon,line_lat = m([filt_csts[x]['Longitude'] for x in csts],[filt_csts[x]['Latitude'] for x in csts])
plt.plot(line_lon,line_lat,'-r')

m.drawcoastlines(linewidth=2)

mpld3.display(fig)
```

```python
cast_line = []
for ii in csts:
    cast_line.append(filt_csts[ii])
```

```python
dpth_gr = np.linspace(5,150,250,endpoint=True)

T = np.zeros((len(cast_line),len(dpth_gr)))
S = np.zeros((len(cast_line),len(dpth_gr)))
sig = np.zeros((len(cast_line),len(dpth_gr)))

for ii in range(len(cast_line)):
    for jj in range(len(dpth_gr)):
        T[ii,jj] = TSmode.get_var_at_depth(cast_line[ii],dpth_gr[jj],'Temperature')
        S[ii,jj] = TSmode.get_var_at_depth(cast_line[ii],dpth_gr[jj],'Salinity')
        sig[ii,jj] = TSmode.get_var_at_depth(cast_line[ii],dpth_gr[jj],'sigmaT')
```

```python
plt.figure(figsize=(16,12))
plt.subplot(3,1,1)
plt.contourf([k['Latitude'] for k in cast_line], -dpth_gr, T.transpose(),cmap='Spectral_r')
plt.colorbar()
for ii in cast_line:
    plt.plot([ii['Latitude'],ii['Latitude']],[min(-dpth_gr),0],'--k',lw=0.5,alpha=0.5)
    
plt.title(r'Temperature')
    
plt.subplot(3,1,2)
plt.contourf([k['Latitude'] for k in cast_line], -dpth_gr, S.transpose(),cmap='Spectral_r')
plt.colorbar()
for ii in cast_line:
    plt.plot([ii['Latitude'],ii['Latitude']],[min(-dpth_gr),0],'--k',lw=0.5,alpha=0.5)
plt.ylabel('Depth')    
plt.title(r'Salinity')

plt.subplot(3,1,3)
plt.contourf([k['Latitude'] for k in cast_line], -dpth_gr, sig.transpose(),cmap='Spectral_r')
plt.colorbar()
for ii in cast_line:
    plt.plot([ii['Latitude'],ii['Latitude']],[min(-dpth_gr),0],'--k',lw=0.5,alpha=0.5)
plt.xlabel('Latitude')
plt.title(r'$\sigma_t$')

plt.show()
```
